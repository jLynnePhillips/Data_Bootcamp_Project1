## HCAHPS Analysis

```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

#### Load the Final Merged Data Set into Data Frame

```python
merged_pd = pd.read_csv("Final_Merged_Data_Set.csv")
merged_pd.head(5)
```

```python
### Box Plot to show the distribution of a HCAHPS score (at a glance)
```

```python
merged_pd.boxplot(['Cleanliness','Nurse_com','Doc_com','Staff_responsiveness','Pain_management','Comm_medicine', 'Overall_Hos_Rating' ],
                  grid=False, rot=90, showmeans= True)
plt.show()
```

```python
mean_scores = {}
mean_scores['Cleanliness']=merged_pd['Cleanliness'].mean()
mean_scores['Nurse_com']=merged_pd['Nurse_com'].mean()
mean_scores['Doc_com'] = merged_pd['Doc_com'].mean()
mean_scores['Overall_Hos_Rating'] =merged_pd['Overall_Hos_Rating'].mean()
mean_scores['Staff_responsiveness']=merged_pd['Staff_responsiveness'].mean()
mean_scores['Pain_management']=merged_pd['Pain_management'].mean()
mean_scores['Comm_medicine']=merged_pd['Comm_medicine'].mean()
mean_scores['Discharge_info']=merged_pd['Discharge_info'].mean()
mean_scores['Care_transition']=merged_pd['Care_transition'].mean()
#mean_scores['Quietness']=merged_pd['Quietness'].mean()
#mean_scores['']
mean_scores_pd = pd.DataFrame(list(mean_scores.items()))
mean_scores_pd
```

# The maximum "Overall HCAHPS Score" per County

```python
merged_grp_county = merged_pd.groupby('County')
maxScore_County = merged_grp_county['Overall_Hos_Rating'].max()
maxScore_County.plot(kind = 'bar', use_index = True, figsize = (15,8))
plt.show()
```
