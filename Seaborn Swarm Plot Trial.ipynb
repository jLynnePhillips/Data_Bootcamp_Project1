```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
merged_pd = pd.read_csv("Final_Merged_Data_Set.csv")
merged_pd.columns
```

```python
mean_scores = {}
mean_scores['Cleanliness']=merged_pd['Cleanliness'].mean()
mean_scores['Nurse_com']=merged_pd['Nurse_com'].mean()
mean_scores['Doc_com'] = merged_pd['Doc_com'].mean()
mean_scores_pd = pd.DataFrame(mean_scores)
mean_scores_pd
```

```python
merged_grp_county = merged_pd.groupby('County')
maxScore_County = merged_grp_county['Overall_Hos_Rating'].max()
maxScore_County.plot(kind = 'bar', use_index = True, figsize = (15,8))
plt.show()
```
