```python
import pandas as pd
import numpy as np
import csv
```

```python
Hospital_merged_pd = pd.read_csv("Hospital_Merged_CleanNoIndex.csv")
```

```python
Hospital_merged_pd.head(5)
```

```python
Hospital_merged_pd = Hospital_merged_pd[Hospital_merged_pd.Overall_Hos_Rating != "Not Available"]
```

```python
Hospital_merged_pd["Overall_Hos_Rating"] = Hospital_merged_pd["Overall_Hos_Rating"].astype("float")
```

```python
County_ratings = Hospital_merged_pd.groupby("County Name")
```

```python
# Calculate Stanadard Deviation of the Overall Hospital Ratings per county.
round(County_ratings["Overall_Hos_Rating"].std(), 2)
```

```python
# Create a Data Frame for the Standard Deviation data
StdDevofRatings_County = pd.DataFrame(round(County_ratings["Overall_Hos_Rating"].std(), 2))
# Drop all NaN values (Standard Devation is NaN because there is only 1 city with data for that county)
#StdDevofRatings_County = StdDevofRatings_County[str(StdDevofRatings_County.Overall_Hos_Rating)!= "NaN"]
#Hospital_merged_pd = Hospital_merged_pd[Hospital_merged_pd.Overall_Hos_Rating != "Not Available"]
StdDevofRatings_County
```

```python
# Convert the Hospital Rating Standard Deviation column into string
StdDevofRatings_County["Overall_Hos_Rating"] = StdDevofRatings_County["Overall_Hos_Rating"].astype("str")
# delete the rows with Standard Devaiation as NaN (after NaN is converted into string it is written as "nan")
StdDevofRatings_County = StdDevofRatings_County[StdDevofRatings_County.Overall_Hos_Rating != "nan"]
StdDevofRatings_County.columns = ["Std_Dev_Overall_Rating"]
StdDevofRatings_County
```

```python
StdDevofRatings_County.to_csv("StdDev_Overall_Ratings_perCounty.csv")
```
