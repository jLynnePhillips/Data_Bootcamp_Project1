```python
import pandas as pd


ca_income = pd.read_csv("2016_income_data.csv")
ca_health = pd.read_csv("2016_health_data.csv")


ca_income_df = pd.DataFrame(ca_income)
ca_health_df = pd.DataFrame(ca_health)


ca_income_df.head()
```

```python
new_header = ca_income_df.iloc[0]
ca_income_df = ca_income_df[1:]
ca_income_df.columns = new_header

ca_income_df.head()
```

```python
list(ca_income_df.columns.values)
```

```python
ca_income_df_cleanup = ca_income_df[["County", "# Deaths", "# Limited Access", "% Limited Access", "# Uninsured", 
                                     "% Uninsured", "Household Income", "Population", "# Rural", "% Rural"]].copy()
ca_income_df_cleanup.head()
```

```python
ca_income_df_cleanup.columns=["County", "# Premature Deaths", "# Child Deaths", "# Infant Deaths", "# Limited Access", 
                             "% Limited Access", "# Uninsured Adults", "# Uninsured Children", "% Uninsured Adults", 
                             "% Uninsured Children", "Household Income", "Population", "# Rural", "% Rural"]
ca_income_df_cleanup.head()
```

```python
ca_health.head()
```

```python
new_header2 = ca_health_df.iloc[0]
ca_health_df = ca_health_df[1:]
ca_health_df.columns = new_header2

ca_health_df.head()
```

```python
list(ca_health_df.columns.values)
```

```python
ca_health_df_cleanup = ca_health_df[["County", "Years of Potential Life Lost Rate", "# Uninsured", "% Uninsured", 
                                     "# Primary Care Physicians", "# Mental Health Providers", "# Medicare Enrollees",
                                    "Preventable Hosp. Rate", "Graduation Rate", "# Some College", "% Some College",
                                    "# Unemployed", "% Unemployed", "# Children in Poverty", "% Children in Poverty",
                                    "80th Percentile Income", "20th Percentile Income", "Income Ratio", "# Single-Parent Households",
                                    "% Single-Parent Households", "# Violent Crimes", "Violent Crime Rate", "# Households with Severe Problems",
                                    "% Severe Housing Problems"]]

ca_health_df_cleanup.head()
```

```python
list(ca_health_df_cleanup.columns.values)
```

```python
ca_health_df_cleanup.columns=["County", "Years of Potential Life Lost Rate", "# Uninsured", "% Uninsured", 
                              "# Primary Care Physicians", "# Mental Health Providers", "Mental Health drop", 
                              "# Medicare Enrollees", "Medicare Enrollees drop","Preventable Hosp. Rate", 
                              "Graduation Rate", "# Some College", "% Some College","# Unemployed", "% Unemployed", 
                              "# Children in Poverty", "% Children in Poverty", "80th Percentile Income", "20th Percentile Income", 
                              "Income Ratio", "# Single-Parent Households", "% Single-Parent Households", "# Violent Crimes", 
                              "Violent Crime Rate", "# Households with Severe Problems", "% Severe Housing Problems"]
ca_health_df_cleanup.head()
```

```python
ca_health_df_drop = ca_health_df_cleanup.drop(["Mental Health drop", "Medicare Enrollees drop",], axis=1)
ca_health_df_drop.head()
```

```python
ca_health_df_drop.to_csv("county_health_df.csv")
ca_income_df_cleanup.to_csv("county_income_df.csv")
```
