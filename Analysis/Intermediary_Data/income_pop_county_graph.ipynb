```python
import pandas as pd

income_county = pd.read_csv("county_income_df.csv")

income_county
```

```python
income_county_clean = income_county.dropna(subset=["County"])
income_county_clean
```

```python
county_name = income_county_clean["County"]
county_name
```

```python
county_income = income_county_clean["Household Income"]

import matplotlib.pyplot as plt

plt.figure(figsize=(25,10))
plt.bar(county_name, county_income)
plt.xticks(rotation=90, fontsize=13)
plt.show()
```

```python
county_pop = income_county_clean["Population"]

plt.figure(figsize=(25,20))
plt.bar(county_name, county_pop)
plt.xticks(rotation=90, fontsize=13)
plt.show()
```
